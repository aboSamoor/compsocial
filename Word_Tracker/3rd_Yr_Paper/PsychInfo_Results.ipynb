```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from os import path
from matplotlib.ticker import ScalarFormatter, FormatStrFormatter
```

```python
df = pd.DataFrame.from_csv("data/PsycInfo/PsycInfo Articles Review.csv")
df2 = pd.DataFrame.from_csv("data/Grants/Processed/nih_budget.csv")
```

```python
x = df.index.year
y = df.values
x2 = df2.index.year
y2 = df2.values
```

```python
y_ = np.log10(y)
m, b = np.polyfit(x, y_, 1)
fig = plt.figure(figsize=(9.0, 6.0))
ax = fig.add_subplot(1,1,1)
_ = plt.plot(x, y,  marker='+', label="Data")
_ = plt.plot(x, 10**(m*x + b), 'r-', label="Best fitting line")
ax.set_yscale('log')
ax.yaxis.set_major_formatter(FormatStrFormatter('%.0f'))
ax.set_ylabel("Articles")
_ = ax.set_xlabel("Year")
plt.tick_params(axis='y', which='minor', labelsize=7)
ax.yaxis.set_minor_formatter(FormatStrFormatter("%.0f"))
_ = ax.legend()
```

```python
fig = plt.figure(figsize=(9.0, 6.0))
ax = fig.add_subplot(1,1,1)
font = {'family': "Times New Roman",
            'color': 'black',
           'size': 12}
y = df.values / 1000.0
_ = plt.plot(x, y,  marker = 'd', markevery =3, linewidth = 2, label="PsycINFO Meta Data", color = 'gray')
y_ = np.log10(y)
m, b = np.polyfit(x, y_, 1)
_ = plt.plot(x, 10**(m*x + b), 'r-', linewidth = 2, label="Curve of Best Fit", color = 'black')
ax.set_ylabel("Number of Publications (in thousands)", font)
_ = ax.set_xlabel("Year", font)
plt.tick_params(axis='y', which='minor', labelsize=7)
#ax.yaxis.set_major_formatter(FixedOrderFormatter(3))
_ = ax.legend(loc='best')
_ = plt.xlim(1845, 2020)
```
