```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
df = pd.DataFrame.from_csv("data/PsycInfo/PsycInfo Articles Review.csv")
```

```python
x = df.index.year
y = df.values
```

```python
y_ = np.log10(y)
m, b = np.polyfit(x, y_, 1)
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(1,1,1)
_ = plt.plot(x, y,  marker='x', label="Data")
_ = plt.plot(x, 10**(m*x + b), 'r.', label="Linear Model")
ax.set_yscale('log')
ax.set_ylabel("Articles")
_ = ax.set_xlabel("Year")
_ = ax.legend()
```
